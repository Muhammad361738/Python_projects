ðŸ”¹ 1. Student Class
Represents a student.

Attributes:

name

roll_number

grade

courses (list)

Methods:

enroll(course)

drop(course)

show_details()



```python
class Student :
    def __init__(self,name,roll_number,grade):
        self.name = name
        self.roll_number = roll_number
        self.grade = grade
        self.courses = []
        
    def enroll(self,course):
        if course in self.courses:
            print("Already Enroll in this course ")
        else:
            self.courses.append(course)
            print(f"Enrolled in {course} successfully!")

        
    def drop(self,course):
        if course not in self.courses:
            print("You are not enroll in this course ")
        else:
            self.courses.remove(course)
            print(f"Dropped {course} successfully.")

        
    def show_details(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.roll_number}")
        print(f"Grade: {self.grade}")
        print(f"Enrolled Courses: {', '.join(self.courses) if self.courses else 'None'}")
```

Manages all students.

Attributes:

students (list of student objects)

Methods:

add_student()

remove_student(roll_number)

find_student(roll_number)

list_all_students()



```python
class Management :
    def __init__(self):
        self.students = []

    def add_student(self,name,roll_number,grade):
        for s in self.students:
            if s.roll_number == roll_number:
                print("Student Alredy Add")
                return
        self.students.append(Student(name,roll_number,grade))
        print("Student Added Successfully ")
        
    def remove_student (self,roll_number):
        for r in self.students:
            if r.roll_number == roll_number:
                self.students.remove(r)
                print("Student Remove Successfully")
                return
        print("Student does not Exist")        
    
    def find_student (self,roll_number):
        for r in self.students:
            if r.roll_number == roll_number:
                r.show_details()
                return
        print("Student does not Exist")        
    
    def list_all_students(self):
        if not self.students:
            print("No Students in Class")
            return
        for s in self.students:
            s.show_details()
            print("--------")
```

```python
# ----------- User Interaction Code -----------
def main ():
    manag = Management()
    while True:
        print(". Student Management System")
        print("1. Add student")
        print("2. Remove student")
        print("3. Enroll in course")
        print("4. Drop course")
        print("5. Show student details")
        print("6. List all students")
        print("7 .Exit")
        
        choice = input("Enter your choice: ")
        if choice == "1":
            name = input("Enter Your Name ")
            roll_number = int(input("Enter Your Roll Number"))
            grade = input("Enter Your Grade")
            
            manag.add_student(name,roll_number,grade)
        elif choice == "2":
            roll_number = int(input("Enter Your Roll Number"))
            manag.remove_student(roll_number)
        elif choice == "3":
            roll_number = int(input("Enter Your Roll Number"))
            student = manag.find_student(roll_number)
            
            if student:
                course = input("Enter course to enroll ")
                student.enroll(course)
                
            else : 
                print("student not found ")
                
        elif choice == "4":
            roll_number = int(input("Enter Your Roll Number"))
            student = manag.find_student(roll_number)
            
            if student:
                 course = input("Enter course to enroll ")
                 student.drop(course)
            else:
                print("Student not found ")
                
        elif choice == "5":
            roll_number = int(input("Enter Your Roll Number"))
            student = manag.find_student(roll_number)
            
            if student:
                student.show_details()
            else:
                print("Student not found")
                
        elif choice == "6":
            manag.list_all_students()
            
        elif choice == "7":
            print("Exiting program.")
            break
        else:
            print("Enter Valid input between 1 to 7")

if __name__ == "__main__":
    main()

```
